Q3.Using the sales data, generate a heatmap to visualize the variation in sales across
different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

```python
# Generate sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Group the data by month and day and calculate the mean sales
sales_heatmap = df.groupby(['Month', 'Day']).mean().reset_index()

# Create a heatmap
fig = px.imshow(
    sales_heatmap.pivot('Month', 'Day', 'Sales').values,
    x=sales_heatmap['Day'].unique(),
    y=sales_heatmap['Month'].unique(),
    labels=dict(x="Day", y="Month", color="Sales"),
    color_continuous_scale='Viridis',
    title='Sales Variation Across Months and Days'
)

# Show the plot
fig.show()
```

Q4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_file('bokeh_histogram.html')
p=figure(title='Histogram')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
show(p)
```
